```python
import pandas as pd
import numpy as np
import math
import random
from sklearn.model_selection import KFold
```

```python
train=pd.read_csv('train.csv')
# train is our training data, toxic is what kind of toxic 
def generate_dataset(train,toxic):
# find toxic is 1
    toxic_comment=train[train[toxic]==1]
    toxic=train[toxic]
# obtain the toxic comment index
    toxic_index=np.where(toxic==1)
    toxic_index=toxic_index[0].tolist()
# drop the toxic comment from the train, only nontoxic remain
    remain_data=train.drop(toxic_index)
    remain=remain_data.index.tolist()
# divided the remain (each size is equal to toxic)
    time=math.ceil(len(remain)/len(toxic_index))
    kf=KFold(time,shuffle=False)
    remain=np.array(remain)
    data_list=[]
    dfsets=[]
    for train_index, text_index in kf.split(remain):  
        X_sample=remain[text_index]
        data_list.append(X_sample)
    datasets=random.sample(data_list,10)
    for data in datasets:
        start=data[0]
        end=data[-1]
        nontoxic=remain_data[start:end]
        df=pd.concat([toxic_comment,nontoxic],axis=0)
        dfsets.append(df)
    return dfsets
```

```python
data=generate_dataset(train,'severe_toxic')
data[0]
```
