```python
import pandas as pd
from sklearn.preprocessing import minmax_scale
```

```python
df0=pd.read_csv('bi_submission.csv')
df1=pd.read_csv('gru_submission.csv')
df2=pd.read_csv('lgb_submission.csv')
df3=pd.read_csv('lr_submission.csv')
#df4=pd.read_csv('submission_lstmcnn_gru_lgb_lr.csv')
df5=pd.read_csv('OOf.csv')
df6=pd.read_csv('tidy_xgb_glm.csv')
```

```python
df_sub1 = pd.DataFrame()
df_sub1["id"] = df0["id"]
df_sub2 = pd.DataFrame()
df_sub2["id"] = df0["id"]
classes = ["toxic", "severe_toxic", "obscene" ,"threat", "insult" ,"identity_hate"]
for cls in classes:
    df_sub1[cls] = (minmax_scale(df0[cls])+minmax_scale(df1[cls]) + minmax_scale(df2[cls]) \
                    + minmax_scale(df3[cls])+minmax_scale(df5[cls])+minmax_scale(df6[cls]))/ 6
    df_sub2[cls] = (minmax_scale(df0[cls]) + minmax_scale(df1[cls]) + minmax_scale(df2[cls]) \
                    + minmax_scale(df3[cls])+minmax_scale(df4[cls])+minmax_scale(df5[cls]))/ 6
```
